```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Juntando dois dataframes diferentes para representar o período de vendas de 2019 por inteiro ##

```python
vendas_df = vendas_df.append(vendas_df_dez)
display(vendas_df)
# Linhas referentes ao DF de Dez integradas ao DF inicial
```

## Começando a analise pelos produtos que estão sendo vendidos ##

```python
produtos = vendas_df[['ID Loja','Produto', 'Quantidade', 'Valor Unitário', 'Valor Final']]
display(produtos)
```

## 1) Qual o produto mais caro e o produto mais barato? ##

```python
produtos_sortmin = produtos.sort_values(['Valor Unitário'])
display(produtos_sortmin)
# par de Meia é o produto mais barato, 30/unidade
# Terno de Linho é o produto mais caro, 750/unidade
```

## 2) Quem vendeu mais e quem vendeu menos meias/ternos de linho?
## 3) Qual foi o total de meias e ternos de linho vendidos no período?

```python
meias_df = produtos.loc[produtos['Produto'] == 'Meia']
meias_grouped = meias_df.groupby('ID Loja')
for key, item in meias_grouped:
    print(meias_grouped.get_group(key).sum(), "\n\n")
# Paladium Shopping Curitiba vendeu mais meias
# Rio Mar Recife vendeu menos meias
```

```python
meias_df[meias_df['Produto'] == 'Meia'].sum()
# O total de meias vendidas em 2019 foi de 1827 unidades
```

```python
linho_df = produtos.loc[produtos['Produto'] == 'Terno Linho']
linho_grouped = linho_df.groupby('ID Loja')
for key, item in linho_grouped:
    print(linho_grouped.get_group(key).sum(), "\n\n")
# Ribeirão Shopping vendeu mais ternos de linho, 97 unidades
# Novo Shopping Ribeirão Preto vendeu menos ternos de linho, 40 unidades
```

```python
linho_df[linho_df['Produto'] == 'Terno Linho'].sum()
# O total de ternos de linho vendidos em 2019 foi de 1741 unidades
```

## 4) Quantos produtos distintos estão sendo vendidos em cada unidade?

```python
produtos_l = vendas_df['Produto'].array
lista_p = []
for i in produtos_l:
    if i not in lista_p:
        lista_p.append(i)
    else:
        continue
print(lista_p)
len(lista_p)

# retorna a lista de produtos sendo vendidos e quantidade de produtos distintos. 120 produtos no total.
```

## 5) Quantas unidades/lojas estão presentes neste estudo e quais foram as 3 unidades com o maior faturamento de 2019?

```python
unidades = vendas_df['ID Loja'].array
lista_u = []
for i in unidades:
    if i not in lista_u:
        lista_u.append(i)
    else:
        continue
print(lista_u)
len(lista_u)

# retorna a lista de unidades de venda e quantidade de unidades. 25 unidades no total.
```

```python
faturamento = vendas_df.groupby('ID Loja')['Valor Final'].sum().sort_values(ascending=False)
display(faturamento)
```

```python
plot_fat = faturamento.reset_index()
# resetando index para poder plotar com base no ID Loja e Valor Final
```

```python
#plot simples mostrando as 5 unidades com o maior faturamento
plt.figure(figsize=(10,5))
ax = sns.lineplot(x='ID Loja',y='Valor Final',data=plot_fat[0:5])
```

```python
faturamento_campinas = vendas_df[vendas_df['ID Loja'] == 'Iguatemi Campinas']
faturamento_campinas_a = faturamento_campinas.groupby('Mês')['Valor Final'].sum().sort_values(ascending=False)
display(faturamento_campinas_a)
```

```python
plot_2019_campinas = faturamento_campinas_a.reset_index()
```

```python
x_campinas = sns.barplot(x='Mês',y='Valor Final',data=plot_2019_campinas,order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
```

```python
# Os meses de maior faturamento na unidade que mais faturou (Iguatemi Campinas) foram os meses de Setembro, Julho e Março.
# Passamos a analisar se o mesmo padrão se repete na unidade com o segundo maior faturamento (Shopping Vila Velha)
```

```python
faturamento_vilavelha = vendas_df[vendas_df['ID Loja'] == 'Shopping Vila Velha']
faturamento_vilavelha_a = faturamento_vilavelha.groupby('Mês')['Valor Final'].sum().sort_values(ascending=False)
display(faturamento_vilavelha_a)
```

```python
plot_2019_vilavelha = faturamento_vilavelha_a.reset_index()
display(plot_2019_vilavelha)
```

```python
x_vilavelha = sns.barplot(x='Mês',y='Valor Final',data=plot_2019_vilavelha,order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
```
