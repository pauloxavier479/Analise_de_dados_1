1 - The first two columns need to be changed to represent date in the right format.

2 - Margem Média Revenda; Preço Médio Distribuição; Desvio Padrão Distribuição; Preço Mínimo Distribuição; Preço Máximo Distribuição; and Coef de Variação Distribuição should be floats.
```python
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"])
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
```

```python
type(df['DATA INICIAL'][0])
```

```python
#Listing all the available products in the dataframe
df["PRODUTO"].value_counts()
```

```python
#To make consulting and plotting gas prices at a given month/year easier, we will use the timestamp column to create new columns with a callable Year, Month and Year-Month format
df["ANO"] = df["DATA INICIAL"].apply(lambda x: '{}'.format(x.year))
df["MÊS"] = df["DATA INICIAL"].apply(lambda x: '{:02d}'.format(x.month))
df["ANO-MÊS"] = df["DATA INICIAL"].apply(lambda x: '{}'.format(x.year)) + df["DATA INICIAL"].apply(lambda x: '-{:02d}'.format(x.month))
```

```python
#And then rearrange the columns to keep the data columns together
df.columns.tolist()
```

```python
#Reorganizing the order of the columns
df = df[[
 'ANO',
 'MÊS',
 'ANO-MÊS',
 'DATA INICIAL',
 'DATA FINAL',
 'REGIÃO',
 'ESTADO',
 'PRODUTO',
 'NÚMERO DE POSTOS PESQUISADOS',
 'UNIDADE DE MEDIDA',
 'PREÇO MÉDIO REVENDA',
 'DESVIO PADRÃO REVENDA',
 'PREÇO MÍNIMO REVENDA',
 'PREÇO MÁXIMO REVENDA',
 'MARGEM MÉDIA REVENDA',
 'COEF DE VARIAÇÃO REVENDA',
 'PREÇO MÉDIO DISTRIBUIÇÃO',
 'DESVIO PADRÃO DISTRIBUIÇÃO',
 'PREÇO MÍNIMO DISTRIBUIÇÃO',
 'PREÇO MÁXIMO DISTRIBUIÇÃO',
 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']]
```

```python
#Also, it gets really annoying having to call each column in all caps all the time. So we're using the lower function to make it more efficient.
df.columns = map(str.lower, df.columns)
```

```python
g = sns.lineplot(data=gc[(gc['estado'] == 'ACRE') | (gc['estado'] == 'SAO PAULO')], x='ano', y='preço médio revenda', hue='estado')
plt.xticks(rotation=45)
```

It's interesting to observe that confronting the mean price of Acre and São Paulo (the states with the highest recorded prices and lowest recorded prices, respectively) we can see that despite the price difference, the price increase followed a very similar pattern.

```python
#We can further explore this idea by checking what was the % increase in these two states over the years
gc_acre = gc[gc["estado"] == 'ACRE'].copy()
gc_acre_ano = gc_acre.groupby('ano')[['preço médio revenda','ano']].mean()
gc_sp = gc[gc["estado"] == 'SAO PAULO'].copy()
gc_sp_ano = gc_sp.groupby('ano')[['preço médio revenda','ano']].mean()
```

```python
var_acre = (gc_acre_ano/gc_acre_ano.shift(1)-1)*100
var_sp = (gc_sp_ano/gc_sp_ano.shift(1)-1)*100
var_sp
sns.lineplot(var_acre, x='ano', y='preço médio revenda',label='Acre').set(title='Gas price % change over the previous year')
sns.lineplot(var_sp, x='ano', y='preço médio revenda',label='São Paulo')
plt.xticks(rotation=45)
```
